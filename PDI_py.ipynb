```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image

# Abrir imagem com OpenCV
img = cv2.imread('sample_data/Imagem.jpg') # l em BGR
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB) # converte para RGB
plt.imshow(img_rgb)
plt.title("Imagem com OpenCV")
plt.axis('off')
plt.show()
```

# Ex. 1

```python
img50 = cv2.resize(img, None, fx=0.5, fy=0.5, interpolation=cv2.INTER_AREA)
img25 = cv2.resize(img, None, fx=0.25, fy=0.25, interpolation=cv2.INTER_AREA)
img10 = cv2.resize(img, None, fx=0.10, fy=0.10, interpolation=cv2.INTER_AREA)

cv2.imwrite("saida_50.jpg", img50)
cv2.imwrite("saida_25.jpg", img25)
cv2.imwrite("saida_10.jpg", img10)
```

# Ex. 2

```python
img_peq = cv2.resize(img, (32, 32), interpolation=cv2.INTER_NEAREST)

img_reamp = cv2.resize(img_peq, (img.shape[1], img.shape[0]), interpolation=cv2.INTER_NEAREST)

cv2.imwrite("saida_amostrada.jpg", img_peq)
cv2.imwrite("saida_reampliada.jpg", img_reamp)
```

# Ex. 3

```python
levels = 4
step = 256 // levels
img_q = (img // step) * step

cv2.imwrite("saida_quantizada.jpg", img_q)
```

# Ex. 4

```python
img_noR = img.copy(); img_noR[:,:,0] = 0
img_noG = img.copy(); img_noG[:,:,1] = 0
img_noB = img.copy(); img_noB[:,:,2] = 0

cv2.imwrite("saida_semR.jpg", img_noR)
cv2.imwrite("saida_semG.jpg", img_noG)
cv2.imwrite("saida_semB.jpg", img_noB)
```
